```python
# importacion general de librerias y de visualizacion (matplotlib y seaborn)
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

plt.style.use('default') # haciendo los graficos un poco mas bonitos en matplotlib

sns.set(style="whitegrid") # seteando tipo de grid en seaborn

pd.options.display.float_format = '{:20,.2f}'.format # suprimimos la notacion cientifica en los outputs
```

```python
apps_details = pd.read_csv('/content/drive/MyDrive/Organización de Datos/TP1 - Visualizaciones/googleplaystore.csv')
```

```python
apps_details
```

```python
#no considero apps duplicadas ni con ratings inválidos
apps_details_rated = apps_details.drop_duplicates(subset='App')
apps_details_rated = apps_details_rated[(apps_details_rated.Rating <= 5)]
```

```python
g2=apps_details_rated['Rating'].plot.hist(figsize=(14,7),bins=15,color='gold', grid=True)
g2.set_title('Distribución de valores del rating de apps', fontsize=18)
g2.set_xlabel('Rating de la app')
g2.set_ylabel('Frecuencia')
```

Se eliminaron los ratings con valor mayor a cinco ya que es imposible que una app pueda tener un rating con esos valores.

La decisión de eliminar apps duplicadas se debe a que las considero un error en el archivo y opté por eliminarlas. Puede verse que si no hacía esto, cambiaba considerablemente el histograma, haciendo crecer algunas barras por la repetición de apps (si una app tenía rating 4.0 y aparecía 10 veces en el csv, estaría contando que el rating 4.0 aparece 9 veces de más de las que en realidad debería).
