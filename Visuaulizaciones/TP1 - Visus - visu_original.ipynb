```python
# importacion general de librerias y de visualizacion (matplotlib y seaborn)
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

plt.style.use('default') # haciendo los graficos un poco mas bonitos en matplotlib

sns.set(style="whitegrid") # seteando tipo de grid en seaborn

pd.options.display.float_format = '{:20,.2f}'.format # suprimimos la notacion cientifica en los outputs
```

```python
apps_details = pd.read_csv('/content/drive/MyDrive/Organizacio패n de Datos/TP1 - Visualizaciones/googleplaystore.csv')
user_reviews = pd.read_csv('/content/drive/MyDrive/Organizacio패n de Datos/TP1 - Visualizaciones/googleplaystore_user_reviews.csv')
```

```python
#no considero apps duplicadas ni con ratings inv치lidos. lo mismo para apps con reviews NaN.
apps_details = apps_details.drop_duplicates(subset='App')
apps_details = apps_details[(apps_details.Rating <= 5)]
user_reviews = user_reviews.dropna()
```

```python
apps_details['Last Update Year']=pd.to_datetime(apps_details['Last Updated']).dt.year
```

```python
apps_details_with_reviews = pd.merge(apps_details, user_reviews)
apps_details_with_reviews = apps_details_with_reviews[['Last Update Year', 'Type', 'Sentiment']]
apps_details_with_reviews['User Reviews'] = True
```

```python
import plotly.express as px
fig=px.treemap(apps_details_with_reviews, path=[px.Constant('User Reviews by apps\' last update year and type (free or paid)'), 'Last Update Year', 'Type', 'Sentiment'], values='User Reviews', width=1350, height=750)
fig.update_traces(root_color='lightgray')
fig.update_traces(marker=dict(cornerradius=3))

fig.update_layout(
    treemapcolorway = ["DeepSkyBlue", "chartreuse", "gold", "red", "LightSkyBlue", "PaleGreen", "LightGoldenRodYellow", "lightcoral"],
    margin = dict(t=50, l=25, r=25, b=25)
) #uso colores del logo de play store, quiz치s no es tan colorblind friendly
fig.show()
```
