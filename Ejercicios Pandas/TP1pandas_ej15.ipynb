# Ejercicio 15
Para aquellas aplicaciones que se encuentren en su versión 1.0 (o 1.0.0) y que tengan más de 50000 reviews. ¿Cuál es el top 5 con mejor rating? (⭐)

Voy a utilizar únicamente las columnas App (nombre de la aplicación), Rating, Reviews (la cantidad de reviews que tiene la aplicación) y Current Ver (la versión de la aplicación).

Los demás detalles sobre las aplicaciones no son útiles para este ejercicio.

```python
apps_details = pd.read_csv('/content/drive/MyDrive/Organización de Datos/googleplaystore.csv', usecols=['App', 'Rating', 'Reviews', 'Current Ver'])
```

Hay apps que figuran repetidas y una tiene rating mayor a 5 (y tiene los otros datos mal cargados) o Nan. No las tengo en cuenta ya las repetidas no me darán información nueva (mismo rating) y las de rating inválido menos.

```python
apps_details = apps_details.drop_duplicates(subset='App')
apps_details = apps_details[(apps_details.Rating <= 5)]
```

```python
apps_details.info() #Reviews es de tipo 'object' (no necesario para resolución)
```

La columna reviews es de tipo object, lo que causará que no pueda filtrar por la cantidad de reviews (no puedo hacer una comparación de algo tipo object con un número). Debido a esto (y a que son números enteros), transformo los datos de esta columna a int.

```python
apps_details['Reviews'] = apps_details['Reviews'].astype(int)
```

```python
apps_details = apps_details[(apps_details.loc[:, 'Current Ver'] == '1.0') | (apps_details.loc[:, 'Current Ver'] == '1.0.0')]
apps_details = apps_details[apps_details['Reviews'] > 50000]
```

```python
apps_details.nlargest(5, "Rating")
```
